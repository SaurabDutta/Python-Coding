```python
selected_rows = df[~df['Out Time ']. isna()]

# ~  is a not operation,it staands for negation. This will print all outtime of employees not having outtime as NaN
```

```python
selected_rows.to_excel("D:\Biometric_play.xlsx")

#Put the filtered output in the Excel file 
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = selected_rows['Employee Name'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('In Time Counts of Week') 
ax.set_xlabel('Employee Name') 
```
